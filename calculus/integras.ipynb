```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(0, 1000 + dt, dt)
a = 0.5
x0 = 2
```

```python
# Linear Ordinary Differential Equation (ODE)
x_analytical = x0 * np.exp(a * t)

sol = solve_ivp(lambda t, x: [a * x[0]], [t[0], t[-1]], [x0], t_eval=t)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(t, x_analytical, '-k', linewidth=2, label='analytical')
ax.plot(sol.t, sol.y[0], '--r', linewidth=2, label='numerical')
ax.set_xlabel('Time'); ax.set_ylabel('Population size')
ax.legend(); ax.set_xlim([0, 20])
plt.tight_layout(); plt.show()
```

```python
# Logistic ODE
b = 0.025
C1 = a / b
C2 = (C1 - x0) / x0

x_analytical = C1 / (1 + C2 * np.exp(-a * t))

sol = solve_ivp(lambda t, x: [a * x[0] - b * x[0]**2], [t[0], t[-1]], [x0], t_eval=t)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(t, x_analytical, '-k', linewidth=2, label='analytical')
ax.plot(sol.t, sol.y[0], '--r', linewidth=2, label='numerical')
ax.set_xlabel('Time'); ax.set_ylabel('Population size')
ax.legend(); ax.set_xlim([0, 20])
plt.tight_layout(); plt.show()
```
