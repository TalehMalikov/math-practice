```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import factorial
```

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)
plt.plot(x, y, 'k', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)

P = [1, 0]
plt.plot(x, np.polyval(P, x), 'b--', linewidth=1.2)

P = [-1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'r--', linewidth=1.2)

P = [1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'g--', linewidth=1.2)

P = [-1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'm--', linewidth=1.2)

P = [1/factorial(9), 0, -1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'c--', linewidth=1.2)

P = [-1/factorial(11), 0, 1/factorial(9), 0, -1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'y--', linewidth=1.2)

P = [1/factorial(13), 0, -1/factorial(11), 0, 1/factorial(9), 0, -1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
plt.plot(x, np.polyval(P, x), 'k--', linewidth=1.2)

plt.legend(['sin(x)', '1st order', '3rd order', '5th order', '7th order', '9th order'])
plt.show()
```
